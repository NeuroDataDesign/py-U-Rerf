```python
import h5py
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
from src.features.summary import Stats
```

```python
public_method_names = [method 
                       for method in dir(Stats) 
                       if callable(getattr(Stats, method)) 
                       if not method.startswith('_')]

public_method_names.remove('return_stats')
```

```python
with h5py.File('sim_results.h5') as hf:
    statistics = hf['statistics'][:]
    time_results = hf['time_results'][:]
```

```python
for i, title in zip(range(2, 26), public_method_names):
    fig = plt.figure()
    ax = plt.axes(projection='3d')

    x = time_results[:,0]
    y = time_results[:,1]
    z = time_results[:,i]
    
    ax.set_title(title)
    
    ax.set_xlabel('num_nodes')
    ax.set_ylabel('P(connection)')
    ax.set_zlabel('runtime (s)')

    ax.scatter(x, y, z)
```
